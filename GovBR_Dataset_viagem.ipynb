```python
df_cargos = df_trip['Cargo'].value_counts().reset_index()
df_cargos
```

```python
df_resumo_top_10 = (
    df_trip.groupby('Cargo')
    .agg(
        despesa_media=('Gasto total', 'mean'),
        duração_media=('Duração da viagem', 'mean'),
        despesas_totais=('Gasto total', 'sum'),
        viagens=('Duração da viagem', 'count'),
        destinos_mais_frequentes=('Destinos', pd.Series.mode)
    )
    .sort_values(by='despesas_totais', ascending=False)
    .head(10)
    .reset_index()
)

df_resumo_top_10['Cargo'] = df_resumo_top_10['Cargo'].str.title()
df_resumo_top_10
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(df_resumo_top_10['Cargo'], df_resumo_top_10['viagens'])
plt.xlabel('Cargo')
plt.ylabel('Número de Viagens')
plt.title('Número de Viagens por Cargo')
plt.xticks(rotation=90)
plt.show()
```

O cargo 'Não Identificado', vindo dos valores NaN, os quais não forampreenchidos adequadamente com todas as informações, ocupa o primeiro lugar em termos de números de viagens por cargo - mais de 23 mil viagens. O segundo maior valor encontra-se em 'Informações Protegidas por Sigilo', que também não se tem uma informação correta, o que corresponde a mais de 9 mil viagens. Os valores caem muito ao longo que os cargos são identificados, apontando mais uma vez à um possível desvio de verba, como os cargos não podem ser identificados, assim como o direcionamento do dinheiro.

```python
import matplotlib.pyplot as plt

# Criar a coluna com a proporção percentual das despesas
df_resumo_top_10['despesa_relativa'] = (
    df_resumo_top_10['despesas_totais'] / df_resumo_top_10['despesas_totais'].sum() * 100
)

# Selecionar os 5 cargos com maiores despesas proporcionais
df_top = df_resumo_top_10.nlargest(5, 'despesa_relativa')

# Grafico
plt.figure(figsize=(9,9))
wedges, texts, autotexts = plt.pie(
    df_top['despesa_relativa'],
    labels=None,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False,
    pctdistance=0.85
)


plt.legend(
    wedges,
    df_top['Cargo'],
    title='Cargo',
    loc='center left',
    bbox_to_anchor=(1, 0.5)
)

plt.title('Top 5 Cargos com Maiores Despesas de Viagens')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

O gráfico pizza acima demonstra os 5 cargos com maiores despesas em viagens - lembre-se de esses são dados nacionais, ou seja, esses são os cinco maiores custos de todos os cargos públicos. O mais chocante é que o cargo 'Não Identificado' ocupa 75% da fatia, e mais de 16% encontra-se em informações sigilosas.

```python
df_trip['Duração da viagem'] = df_trip['Duração da viagem'].dt.total_seconds() / 86400
```

```python
df_resumo_top_10 = (
    df_trip.groupby('Cargo')
    .agg(
        despesa_media=('Gasto total', 'mean'),
        duração_media=('Duração da viagem', 'mean'),
        despesas_totais=('Gasto total', 'sum'),
        viagens=('Duração da viagem', 'count'),
        destinos_mais_frequentes=('Destinos', pd.Series.mode)
    )
    .sort_values(by='despesas_totais', ascending=False)
    .head(10)
    .reset_index()
)

df_resumo_top_10['Cargo'] = df_resumo_top_10['Cargo'].str.title()
```

```python
df_resumo_top_10 = df_resumo_top_10.sort_values(by='duração_media', ascending=False)
plt.figure(figsize=(10, 6))
plt.barh(df_resumo_top_10['Cargo'], df_resumo_top_10['duração_media'], color='skyblue')
plt.xlabel('Duração média (dias)')
plt.ylabel('Cargo')
plt.title('Duração média das viagens por cargo')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

O gráfico acima demonstra a duração média das viagens por cargo dentro do período de um ano (365 dias). Em primeiro lugar, encontra-se 'Primeiro Sargento', com uma duração de 217 dias de viagem ao longo de 365 dias. Para maiores conclusões quanto à veracidade desses dados, há necessidade de maiores investigações. Contudo, o apontamento principal seria a comparação de gastos totais: em 217 dias de viagem, a despesa total foi de 2 milhões de reais. Em contrapartida, o cargo 'Não Identificado' teve uma duração de 26 dias de viagens, com uma despesa total de 222 milhões de reais.
